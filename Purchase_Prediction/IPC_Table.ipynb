```python
import pandas as pd
```

```python
ipc = pd.read_html('https://www.ine.es/prensa/ipc_tabla.htm')[1]
```

```python
ipc['Tasa índice general']['Mensual (en %)']
```

```python
df_ipc = pd.DataFrame()
```

```python
df_ipc['Periodo'] = ipc['Período']
```

```python
df_ipc['IPC'] = ipc['Tasa índice general']['Mensual (en %)']
```

```python
ipc
```

```python
df_ipc.dtypes
```

```python
diccionario_mes = {'Diciembre':12, 'Noviembre':11, 'Octubre':10 ,'Septiembre': 9, 'Agosto':8, 'Julio':7, 'Junio':6, 'Mayo':5, 'Abril': 4, 'Marzo': 3, 'Febrero': 2, 'Enero': 1}
```

```python
diccionario_mes
```

```python
df_ipc['mes'] = df_ipc['Periodo'].apply(lambda x: x.split()[0])
```

```python
df_ipc['year'] = df_ipc['Periodo'].apply(lambda x: x.split()[1])
```

```python
df_ipc
```

```python
df_ipc['mes'] = df_ipc['mes'].apply(lambda x: diccionario_mes[x])
```

```python
diccionario_mes['Diciembre']
```

```python
df_ipc['date']  = pd.to_datetime(df_ipc.year.astype(str) + '/' + df_ipc.mes.astype(str) + '/28')
```

```python
df_ipc.drop(['Periodo','mes','year'],axis = 1,inplace=True)
```
